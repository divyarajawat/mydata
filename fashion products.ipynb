# Step 2 : Data Cleaning and Transformation:

```python
df.dropna()
```

```python
df=df[['User ID', 'Product ID', 'Product Name', 'Brand', 'Category', 'Price',
       'Rating', 'Color', 'Size']].copy()
```

```python
df.drop(['Rating'], axis=1)
```

```python
#Fills missing values with a specific value.
df.fillna('Brand') \
.head(10)
```

```python
df.duplicated() # Returns a boolean Series indicating duplicate rows.
```

```python
df.drop_duplicates()  #Drops duplicate rows from the DataFrame.
```

```python
df=df.rename(columns={'Category':'Product_Category'})  #Renames columns in the DataFrame.
```

```python
 #Replaces specific values in the DataFrame.

df.replace('XL', 'XXL') \
   .head(10)
```

```python
df.isna().sum()
```

```python
df.loc[df.duplicated()]
```

```python
df.loc[df.duplicated(subset=['Brand'])]
```

```python
#checking a example for duplicate
df.query('Brand=="Adidas"')
```

```python
df=df.loc[~df.duplicated(subset=['Brand','Rating','Color'])] \
    .reset_index(drop=True).copy()
```
